```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Data Sets/drive-download-20230821T082220Z-001.zip (Unzipped Files)/LoanApprovalPrediction.csv")
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
cols=['Gender','Married','Dependents','Self_Employed','Loan_Amount_Term' ,'Credit_History']
for col in cols:
  df[col]=df[col].fillna(df[col].mode()[0])
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
df.isna().sum()
```

```python
#map

df['Gender']=df['Gender'].map({'Male':0,'Female':1})
df
```

```python
df['Married']=df['Married'].map({'Yes':1,'No':0})
df
```

```python
df.dtypes
```

```python
df['Dependents'].value_counts()
```

```python
df['Dependents']=df['Dependents'].map({'0':0,"1":1,'2':2,'3+':3})
df
```

```python
df['Education'].value_counts()
```

```python
df['Education']=df['Education'].map({'Graduate':1,'Not Graduate':0})
df
```

```python
#self_employed
#yes=1
#no=0
df['Self_Employed']=df['Self_Employed'].map({'Yes':1,'No':0})
df
```

```python
df['Property_Area'].value_counts()
```

```python
df['Property_Area']=df['Property_Area'].map({'Rural':0,'Urban':1,'Semiurban':2})
df
```

```python
df['Loan_Status']=df['Loan_Status'].map({'Y':1,'N':0})
df
```

```python
df.isna().sum()
```

```python
X=df.iloc[:,1:-1]
X
```

```python
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,random_state=1,test_size=0.3)
```

```python
#SVC
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train,y_train)
```

```python
y_pred=sv.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
print(accuracy_score(y_test,y_pred)*100)
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
```
